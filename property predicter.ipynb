```python
# Wrangle function to clean DataFrame
def wrangle( ):
    
    # Read CSV file
    df = pd.read_csv('property.csv')  #Reading directly, I will have raw data from one file

    # Subset data: Lets look for property with bedrooms
    #mask_property_vacant = df["Name"].str.contains("Vacant") #For testing Vacant
    mask_property_beds = df["Name"].str.contains("Bed") | df["Name"].str.contains("arm") | df["Name"].str.contains("acant")
    
    #There must be price
    #mask_price = df["Price"].notnull()

    #I want properties that have size in m²
    mask_size_metres = df["Size"].str.contains("m²")
    
    df = df[mask_size_metres & mask_property_beds]

    #Clean the Size
    df["Size"] = df["Size"].astype(str).str.strip()

    #I wont be needing these columns
    df.drop(columns=["Unnamed: 0", "Location"], inplace=True)
    
    return df
```

```python
#Now lets remove the unecessary tetxt on the data
#Prepare the data so we can build a model
import re

def cleaning(data):
    
    data.insert(loc= 0, column= "Type", value = "")
    #Adding Bed room number, By extracting only digits
    data["Beds"] = data["Name"].str.extract('(\d+)')
    
    #Let me add apartment type
    data.loc[data["Name"].str.contains("part"),"Type"] = "Apartment"               #Apartment or apartment
    data.loc[data["Name"].str.contains("ouse"),"Type"] = "House"                   #h[ouse]
    data.loc[data["Name"].str.contains("arm"),"Type"] = "Farm"                     #F[arm]
    data.loc[data["Name"].str.contains("acant"),"Type"] = "Vacant land"
    #[df["Name"].str.contains("Apart") | df["Name"].str.contains("House")]
    
    #AND DROP THE NAME COLUMN
    data.drop(columns=["Name"], inplace=True)
    
    #Remove unecessary data
    data["Price"] = data["Price"].astype(str).str.replace("R", "", regex=False)
    data["Size"] = data["Size"].astype(str).str.replace(" m²", "", regex=False)

    #Remove white spaces
    data["Price"] = data["Price"].str.replace(" ", "")
    data["Price"] = data["Price"].str.replace("\n", "")
    data["Size"] = data["Size"].str.replace(" ", "")
    data["Size"] = data["Size"].str.replace("\n", "")
    
    #Convert to appropriate data types
    data["Price"] = data["Price"].str.extract('(\d+)')  #Emergency
    data["Size"] = data["Size"].str.extract('(\d+)')  #Emergency
    
    data["Price"] = data["Price"].astype(float) #There is no float at the time
    data["Size"] = data["Size"].astype(int)
    data["Type"] = data["Type"].astype(str)

    # Subset data: Remove outliers for "surface_covered_in_m2" trim by 5%
    low, high = data["Size"].quantile([0.1, 0.9])
    mask_area = data["Size"].between(low, high)
    data = data[mask_area]
    
    return data
```
